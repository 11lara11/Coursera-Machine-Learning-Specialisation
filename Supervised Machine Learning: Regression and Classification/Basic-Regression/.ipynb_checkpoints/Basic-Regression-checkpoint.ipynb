# Notebook covering a simple linear regression model

```python
# import
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv
from sklearn.linear_model import LinearRegression
```

```python
# Find the current working directory. Use ! to run this bash command from within the notebook
!pwd
```

```python
# Load the data using pandas csv_read since it is faster than an equivelant numpy load
RegressionTrainDF = read_csv('Data Sets/Linear Regression Data Set 1 Train.csv')
RegressionTestDF = read_csv('Data Sets/Linear Regression Data Set 1 Test.csv')

# Then convert the data frame to a numpy array
RegressionTrain = RegressionTrainDF.to_numpy()
RegressionTest = RegressionTestDF.to_numpy()
RegressionTrainHeaders = RegressionTrainDF.columns.to_numpy()

# Split the data into the X and Y data sets and shape them to be 2D since Scikit-Learn's LinearRegression requires 2D arrays as input
xTrain = RegressionTrain[:,0].reshape(-1, 1)
yTrain = RegressionTrain[:,1].reshape(-1, 1)
xTest = RegressionTest[:,0].reshape(-1, 1)
yTest = RegressionTest[:,1].reshape(-1, 1)

# Take a look at the data
print(RegressionTrainHeaders,'\n')
print(xTrain[:5],'\n')
print(yTrain[:5])
```

```python
plt.scatter(xTrain, yTrain, marker='x', c='r')
plt.scatter(xTest, yTest, marker='.', c='b')
plt.title("X vs Y")
plt.xlabel(RegressionTrainHeaders[0])
plt.ylabel(RegressionTrainHeaders[1])
```

```python
# Now I want to move on to training a linear regression model. 
# Create an instance of a linear regression model and fit it to the data with the fit() function:
model = LinearRegression()
model.fit(xTrain, yTrain) 
```

```python
# The following section will get results by interpreting the created instance: 
# Obtain the coefficient of determination (1 is perfect):
r_sq = model.score(xTrain, yTrain)
print('coefficient of determination:', r_sq)

# Print the Intercept:
ModelIntercept = 'intercept:', model.intercept_
print('Intercept: ', ModelIntercept)

# Print the Slope:
ModelSlope = model.coef_
print('slope:', ModelSlope) 

# Use the model on the test set to make predictions
Predictions = model.predict(xTest)
```

```python
# Now add the predictions to the original plot
plt.scatter(xTrain, yTrain, marker='x', c='r')
plt.scatter(xTest, yTest, marker='.', c='b')
plt.title("X vs Y")
plt.xlabel(RegressionTrainHeaders[0])
plt.ylabel(RegressionTrainHeaders[1])
plt.scatter(xTest, Predictions, marker='*', c='y')
```
